```python
def solved(crossword):
    empty_spaces = 0
    for line in crossword:
        empties = sum([1 for s in line if s == "-"])
        empty_spaces += empties
    return empty_spaces == 0

def getMarks(crossword):
    marks = {}
    for row in range(10):
        for col in range(10):
            if crossword[row][col] == "-":
                if (col == 0 or crossword[row][col-1] == '+') and col < 9 and crossword[row][col+1] == "-":
                    marks[(row, col)] = [0, 0]
                    i = col
                    while i <= 9 and crossword[row][i] != '+':
                        i += 1
                    marks[(row, col)][1] = i - col
                elif (row == 0 or crossword[row-1][col] == '+') and row < 9 and crossword[row+1][col] == "-":
                    marks[(row, col)] = [1, 0]
                    i = row
                    while i <= 9 and crossword[i][col] != '+':
                        i += 1
                    marks[(row, col)][1] = i - row
    return marks

def isValid(section, word):
    for i in range(len(word)):
        if section[i] != "-" and section[i] != word[i]:
            return False
    return True

def fit(crossword, marks, word):
    crossword_copy = crossword.copy()
    success = True
    for (row, col),[d, l] in marks.items():
        if l == len(word) and (crossword_copy[row][col] == "-" or crossword_copy[row][col] == word[0]):
            if d == 0:
                cross_row = crossword_copy[row]
                if isValid(cross_row[col:col+l], word):
                    cross_row = cross_row[:col] + word + cross_row[col+l:]
                    crossword_copy[row] = cross_row
                else:
                    success = False
                    break
                break
            else:
                for i in range(l):
                    cross_row = crossword_copy[row+i]
                    if cross_row[col] == "-":
                        cross_row = cross_row[:col] + word[i] + cross_row[col+1:]
                        crossword_copy[row+i] = cross_row
                    elif cross_row[col] != word[i]:
                        success = False
                        break
                break
    return crossword_copy if success else crossword
    
def crosswordPuzzleRec(crossword, marks, words):
    if not words:
        return crossword
    
    for i in range(10):
        print(crossword[i])
    print("\n\n")
    
    for i in range(len(words)):
        new_fit = fit(crossword, marks, words[i])
        res = crosswordPuzzleRec(new_fit, marks, words[:i] + words[i+1:])
        if solved(res):
            return res
    return crossword

def crosswordPuzzle(crossword, words):
    marks = getMarks(crossword)
    return crosswordPuzzleRec(crossword, marks, words)
```

### MAX XOR

```python
def maxXor(arr, queries):
    maxes = []
    trie = {}
    k = len(bin(max(arr + queries))) - 2
    
    for number in ['{:b}'.format(x).zfill(k) for x in arr]:
        node = trie
        for char in number:
            node = node.setdefault(char, {})
    
    for n in queries:
        node = trie
        s = ''
        for char in '{:b}'.format(n).zfill(k):
            tmp = str(int(char) ^ 1)
            tmp = tmp if tmp in node else char
            s += tmp
            node = node[tmp]
        maxes.append(int(s, 2) ^ n)
    return maxes
```

### DFS: Connected Cell in a Grid

```python
def maxRegion(grid):
    # Write your code here
    rows, cols = len(grid), len(grid[0])
    max_region = 0
    visited = [[0 for _ in range(cols)] for _ in range(rows)]
    neighbors = [[1, 0], [1, -1], [1, 1], [-1, 0], [-1, -1], [-1, 1], [0, -1], [0, 1]]
    print(visited)
    for i in range(rows):
        for j in range(cols):
            # DFS
            if grid[i][j] and not visited[i][j]:
                size_curr_region = 0
                stack = [(i, j)]
                while stack:
                    print(stack)
                    a,b = stack.pop()
                    visited[a][b] = 1
                    size_curr_region += 1
                    for [na, nb] in neighbors:
                        if a+na >= 0 and a+na < rows and b+nb >= 0 and b+nb < cols:
                            if grid[a+na][b+nb] == 1 and not visited[a+na][b+nb]:
                                stack.append((a+na, b+nb))
                                visited[a+na][b+nb] = 1  
                max_region = max(max_region, size_curr_region)
    return max_region
```

```python
def find(Parents, q):
    if q not in Parents:
        Parents[q] = [q, 1]
        return [q, 1]
    elif Parents[q][0] == q:
        return Parents[q]
    else:
        result = find(Parents, Parents[q][0])
        Parents[q] = result
        return result
        
# Complete the maxCircle function below.
def union(Parents, a, b):
    a_rep = find(Parents, a)
    b_rep = find(Parents, b)
    
    if a_rep[0] == b_rep[0]:
        return a_rep[1]
        
    new_size = a_rep[1] + b_rep[1]
    Parents[a_rep[0]] = [b_rep[0], new_size]
    Parents[b_rep[0]][1] = new_size
    return new_size

def maxCircle(n, c_lib, c_road, queries):
    Parents = {}
    visited = [0] * n
    components = []

    for a, b in queries:
        new_size = union(Parents, a, b)
        #max_size = max(max_size, new_size)     
    
    for node in Parents.keys():
        if visited[node-1] or visited[Parents[node][0]-1]:
            continue
        aux = node
        while True:
            visited[aux-1] = 1
            if Parents[aux][0] == aux:
                components.append(Parents[aux][1])
                break
            aux = Parents[aux][0]
    print(Parents)
    print(components)
    missed = 0
    for i in range(1, n+1):
        if not Parents.get(i, False):
            missed += 1
    missed *= c_lib
    cost_roads = sum([c_road * (c - 1) for c in components]) + c_lib * len(components)
    cost_libs = c_lib * n
    return min(cost_roads, cost_libs) + missed
```

```python
def roadsAndLibraries(n, c_lib, c_road, cities):
    # Write your code here
    neighbors = {(i+1):[] for i in range(n)}
    visited = [0] * n
    components = []
    
    for c in cities:
        print(c)
        if c[1] not in neighbors[c[0]]:
            neighbors[c[0]].append(c[1])
            neighbors[c[1]].append(c[0])
    
    queue = []
    visited_count = 0
    component_size = 0
    
    for node in range(1, n + 1):
        if not visited[node-1]:
            queue.append(node)
            visited[node-1] = 1
            while queue:

                curr = queue.pop(0)
                component_size += 1
                for neigh in neighbors[curr]:
                    if not visited[neigh-1]:
                        visited[neigh-1] = 1
                        queue.append(neigh)
            components.append(component_size)
            component_size = 0

    cost = [c_road * (c - 1) + c_lib for c in components]
    return min(sum(cost), n * c_lib)
```

```python
def getMinimumCost(k, c):
    c = sorted(c)
    min_price = 0
    for i in range(len(c) - 1, -1, -1):
        min_price += c[i] * ((len(c) - i - 1) // k + 1)
    return min_price
```

```python
def sherlockAndAnagrams(s):
    substrings = {}
    for i in range(len(s)):
        for j in range(1, len(s) - i + 1):
            #print(s[i:i+j])
            sorted_sub = "".join(sorted(s[i:i+j]))
            if not substrings.get(sorted_sub, False):
                substrings[sorted_sub] = 1
            else:
                substrings[sorted_sub] += 1
    #print(substrings)
    anagramic_pairs = 0
    for v in substrings.values():
        if v > 1:
            anagramic_pairs += v*(v-1)//2
        
    return anagramic_pairs
```
